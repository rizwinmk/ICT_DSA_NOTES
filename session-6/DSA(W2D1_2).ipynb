```python
price=[]
price.append(int(input("enter the price of first item")))
price.append(int(input("enter the price of second item")))
price.append(int(input("enter the price of third item")))
price.append(int(input("enter the price of fourth item")))
price.append(int(input("enter the price of fifth item")))

total=sum(price)

if total >= 2000 :
  final_bill=total-((20*total)/100)
  discount = 20
elif total >=1000 and total < 2000:
  final_bill=total-((10*total)/100)
  discount = 10
else:
  final_bill=total

print(f"Your total bill amount is {total}Rs.\nYou have availed a discount of {discount}%.\nYou final bill amount is {final_bill}Rs.")
```
