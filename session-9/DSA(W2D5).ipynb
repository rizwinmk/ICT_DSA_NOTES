```python
class bankaccount:
  def __init__(self,owner,account,balance):
    self.owner=owner
    self.account=account
    self.balance=balance

  def deposit(self,amount):
   if amount>0:
    print("incorrect amount")
   else:
    self.balance+=amount
    print(f"Deposited {amount}rs to your bank account.\nAvailable balance: {self.balance}")


  def withdraw(self,amount):
   if self.balance>=amount:
     self.balance-=amount
     print(f"withdrawn {amount}rs from your bank account.\nAvailable balance: {self.balance}")
   else:
    print("insufficient balance")


  def display(self):
    print(f"owner name: {self.owner}\nAccount no: {self.account}\nBalance: {self.balance}")




```

```python
obj1=bankaccount("ashish","AS101",24000)
obj2=bankaccount("Aslaf","AS102",30000)
obj2.deposit(500)
obj1.deposit(1000)
obj1.withdraw(400)
obj1.display()
obj1.withdraw(24601)
```

```python
class savings(bankaccount):
  def __init__(self,owner,account,balance,intrest_rate,min_bal):
    super().__init__(owner,account,balance)
    self.intrest_rate=intrest_rate
    self.min_bal=min_bal

  def cal_intrest(self):
    intrest=self.balance*self.intrest_rate
    self.balance+=intrest

  def withdraw(self,amount):
    if self.balance - amount>=self.min_bal:
      super().withdraw(amount)
    else:
      print("insufficient money")
```

```python
obj11=savings("ashish","AS101",24500,5,1000)
obj11.withdraw(100)
obj11.cal_intrest()
```
