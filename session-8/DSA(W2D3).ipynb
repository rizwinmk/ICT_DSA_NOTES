***Iterative Statements(Loops)***

```python
repeat=int(input("enter the number of times you want your name to repeat: "))
for i in range(repeat):
  print("Ashish")
```

```python
repeat=int(input("enter the number of times you want your name to repeat: "))
i=1
while i<=repeat:
  print("Ashish")
  i=i+1
```

```python
repeat=int(input("enter the number of times you want your name to repeat: "))
i=1
while repeat>=i:
  print(f"Ashish {repeat-i+1}")
  repeat=repeat-1
```
